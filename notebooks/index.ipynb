### 1.2 Text Transformations

In this section, we will prepare the data dataframe for `EDA` and `Modeling`.

- We will drop columns that have no analytical values like description, verified_purchase etc.

The 2 columns have been joined into one column. This will ensure we get the full context of the reviewer's sentiment

```python
# Drop columns that have no anlytical and modeling use

data = data.drop(columns=['verified_purchase','description','rating_desc','average_rating','review_text'])

data.columns
```

```python
# Reorder columns
columns_order = ['rating', 'timestamp', 'movie_title', 'price',
       'categories','reviews','sentiment']
data = data[columns_order]

print(data.columns)
```

```python
data.head(3)
```

Columns are reordered and will proceed to tokenize the reviews column

```python
# Remove any html tags on the reviews text
def strip_html_tags(text):
    soup = BeautifulSoup(text, "html.parser")
    stripped_text = soup.get_text()
    return stripped_text
```

```python
data['reviews'] = data['reviews'].apply(strip_html_tags)
data['reviews'][:3]
```

```python
# Define a function for text preprocessing
def preprocess_text(text):
    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Remove punctuation and symbols
    text = re.sub(r'[^\w\s]', '', text)

    # Convert to lowercase
    text = text.lower()

    # Lemmatize text
    lemmatizer = WordNetLemmatizer()
    text = ' '.join([lemmatizer.lemmatize(word) for word in text.split()])

    # Remove stop words
    stop_words = set(stopwords.words('english'))

    text = ' '.join([word for word in text.split() if word not in stop_words])

    return text
```

```python
# Apply the preprocessing function to the 'reviews'
data["clean_review"] = data["reviews"].apply(preprocess_text)

# verify the changes

data['clean_review'].head()
```

The 'reviews' column has now been transformed into lowercase strings without numbers and symbols. The next step is to tokenize the cleaned text. Tokenization is the process of converting the tweets into individual words. This facilitates better analysis of the words.

```python
# Define a function for tokenization
def tokenize_text(text):
    # Tokenize the text
    tokens = word_tokenize(text)
    return tokens
```

```python
# Create a new column for the tokenized text
data['reviews_tokenized'] = data['reviews'].apply(tokenize_text)

data['reviews_tokenized'].head(3)
```

The reviews have been tokenized and we can now use these words to perform various EDA tasks.

### 2.1.7 Top N-grams

N-grams, including trigrams (3-grams), are critical in Natural Language Processing (NLP) as they help capture context and relationships between words in text. Here's why they matter:

`Context Preservation`: N-grams consider sequences of words, allowing models to understand dependencies and improve predictions.

`Improved Features`: They enhance text representations by capturing multi-word phrases (e.g., `artificial intelligence`), which are often more meaningful than individual words.

`Sentiment and Semantic Analysis`: Trigrams provide richer features for detecting subtleties, such as negations or idiomatic expressions (e.g., `not very good`).


#### Bigrams

```python
# Create a function to get n-grams
def get_top_ngrams(text, ngram_range=(2,2), n=10):
    vectorizer = CountVectorizer(ngram_range=ngram_range, stop_words="english")
    ngram_counts = vectorizer.fit_transform(text)
    ngram_sum = np.asarray(ngram_counts.sum(axis=0)).flatten()

    ngrams_freq = list(zip(vectorizer.get_feature_names_out(), ngram_sum))
    ngrams_freq = sorted(ngrams_freq, key=lambda x: x[1], reverse=True)[:n]

    return ngrams_freq
```

### Extract Most Common Adverbs & Adjectives

```python
def extract_words(text, pos_type):
    doc = nlp(text)
    words = [token.text.lower() for token in doc if token.pos_ == pos_type and not token.is_stop]
    return words
```

```python
# Function to extract POS words for a specific sentiment
def get_pos_by_sentiment(data, sentiment_label, pos_type):
    filtered_texts = data[data["sentiment"] == sentiment_label]["reviews"]
    words = []
    filtered_texts.apply(lambda x: words.extend(extract_words(x, pos_type)))
    return Counter(words).most_common(20)  # Return top 20 most common words
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

def plot_pos_words_comparison(data_dict, pos_type, figsize=(12, 8), color_palette="Set2"):
    """
    Plots a grouped bar chart to compare POS word frequencies across sentiments.

    Args:
    - data_dict (dict): Dictionary where keys are sentiment labels (e.g., "positive")
      and values are lists of (word, count) from the Counter object.
    - pos_type (str): The type of POS tags analyzed (e.g., "ADJ" or "VERB").
    - figsize (tuple): Size of the figure (default: (12, 8)).
    - color_palette (str): Color scheme for the chart (default: "Set2").
    """
    # Flatten the data into a DataFrame for plotting
    data = []
    for sentiment_label, words_counts in data_dict.items():
        for word, count in words_counts:
            data.append({"Word": word, "Frequency": count, "Sentiment": sentiment_label})

    df = pd.DataFrame(data)

    # Create grouped bar chart
    plt.figure(figsize=figsize)
    sns.barplot(data=df, x="Word", y="Frequency", hue="Sentiment", palette=color_palette)
    plt.xlabel('Words', fontsize=14)
    plt.ylabel('Frequency', fontsize=14)
    plt.title(f'Comparison of Top {pos_type} Across Sentiments', fontsize=16)
    plt.xticks(rotation=45, fontsize=12)
    plt.legend(title="Sentiment")
    plt.tight_layout()  # Adjust layout to avoid clipping
    plt.show()
```

The training set will be used to fit the model and learn patterns from the data; we will use the **validation set** to tune hyperparameters (learning rate, number of layers etc) and evaluate the model during training. - We will stratify the dataset to ensure classdistribution in each split remains similar to the orignal dataset.

- Validation set is set at 25% of the remaining 80% ,resulting in a finalsplit of 60% training,20% validation and 20% test.
- Random state ensures reprodicibility so that we can get the same splits each time.

### 3.3 Hyperparameter Tuned SVM Model

Using GridSearchCV, exprole a variety of parameter combinations to optimize class separation and address the bias towards the positive class

The tuning was done in another model and the best params were identified. We are using the best params below to save on computational time.

```python
# Define param grid
param_grid = {
    #'svm__C': [0.1, 1, 10, 100], # Regularization strength
    'svm__C': [10],
    #'svm__kernel': ['linear', 'rbf'],         # Types of kernels
    'svm__kernel': ['rbf'],
    #'svm__gamma': ['scale', 'auto', 0.01, 0.1, 1],  # Kernel coefficient for non-linear kernels
    'svm__gamma': [0.01],
    #'svm__class_weight': ['balanced', None],  # Handling class imbalance
    'svm__class_weight': ['balanced']
}

# Define the GridSearchCV
grid_search = GridSearchCV(
    svm_pipeline,
    param_grid=param_grid,
    scoring='recall_macro',  # Focus on improving macro-averaged recall
    cv=5,                    # 5-fold cross-validation
    n_jobs=-1                # Use all CPU cores for parallel processing
)

# Fit the grid search to the data
grid_search.fit(X_train, y_train)

# Best parameters and score
print("Best Parameters:", grid_search.best_params_)
print("Best Macro Recall:", grid_search.best_score_)

```

### 3.4 Class Balanced Tuned SVM Model

```python
# Separate the classes
negative_class = data[data['sentiment'] == 'negative']
neutral_class = data[data['sentiment'] == 'neutral']
positive_class = data[data['sentiment'] == 'positive']

# Check class sizes
print("Class sizes before undersampling:")
print(f"Negative: {len(negative_class)}, Neutral: {len(neutral_class)}, Positive: {len(positive_class)}")

# Undersample the positive class to match the size of the neutral class
positive_class_undersampled = resample(positive_class,
                                       replace=False,  # Sampling without replacement
                                       n_samples=4000,  # Undersample the Positive Class
                                       random_state=42)  # For reproducibility

# Combine the balanced classes
balanced_data = pd.concat([negative_class, neutral_class, positive_class_undersampled])

# Shuffle the dataset
balanced_data = balanced_data.sample(frac=1, random_state=42).reset_index(drop=True)

# Check class sizes after undersampling
print("Class sizes after undersampling:")
print(balanced_data['sentiment'].value_counts())

# Split into features and target
X_balanced = balanced_data['clean_review']
y_balanced = balanced_data['sentiment']

# Re-train the fine-tuned SVM model
best_svm_pipeline.fit(X_balanced, y_balanced)

# Evaluate the re-trained model
y_val_pred_balanced = best_svm_pipeline.predict(X_val)

# Classification report and confusion matrix
# Classification report
print("Classification Report (After Undersampling):\n")
print(classification_report(y_val, y_val_pred_balanced))

# Confusion matrix
balanced_cm = confusion_matrix(y_val, y_val_pred_balanced)
plt.figure(figsize=(8, 6))
sns.heatmap(balanced_cm, annot=True, fmt='d', cmap='Blues', xticklabels=best_svm_pipeline.classes_, yticklabels=best_svm_pipeline.classes_)
plt.title("Confusion Matrix Hyperparameter Tuned Balanced SVM Model ")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.savefig("Confusion_Matrix_Balanced_SVM.png")
plt.show()
```

**Positive Class**
- Precision: Slight improvement (`0.97 to 0.98`) after undersampling, meaning there were fewer false positives.
- Recall: Decreased slightly (`0.84 to 0.82`), indicating that the model captured slightly fewer true positives from the positive class after undersampling.
- The undersampling caused a small trade-off in recall but improved precision.
- F1-score: A minor drop (`0.90 to 0.89`), reflecting a balance between precision and recall.

**Neutral Class**
- Precision: Improved marginally (`0.43 to 0.46`).
- Recall: Significant improvement (`0.67 to 0.79`), meaning the model performed better at identifying true neutral instances post-undersampling, a significant boost in recall.
- F1-score: Noticeable improvement (`0.53 to 0.58`).

**Negative Class**
- Precision: Slight improvement (0.63 to 0.64).
- Recall: Improved significantly (from 0.81 to 0.88), showing better identification of true negatives after undersampling.
- F1-score: A minor improvement (from 0.71 to 0.74).

Overall, while accuracy stayed constant at 82%, the macro average scores improved, highlighting better balance across classes. This is especially critical when addressing class imbalance.

The undersampling approach prioritized fairness across classes, slightly sacrificing performance on the majority class (positive) for better results in the minority classes.
